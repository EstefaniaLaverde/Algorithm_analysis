```python
## THIS ARBITRARY GRAPH CREATED BY GPT

nodes = [0, 1, 2, 3]
edges = [
    (0, 1, 1),  # Edge from 0 to 1 with weight 1
    (1, 2, 3),  # Edge from 1 to 2 with weight 3
    (0, 2, 2),  # Edge from 0 to 2 with weight 2
    (2, 3, 4),  # Edge from 2 to 3 with weight 4
    (1, 3, 5)   # Edge from 1 to 3 with weight 5
]
    
graph = Graph(nodes, edges)
    
mst = graph.kruskal()
    
print("Minimum Spanning Tree (MST):")
for u, v, w in mst:
    print(f"Edge ({u}, {v}) with weight {w}")

graph.draw_digraph()

solution_graph = Graph(nodes, mst)
solution_graph.draw_graph()
```

```python
## THIS ARBITRARY GRAPH CREATED BY GPT

nodes = [0, 1, 2, 3, 4, 5]
edges = [
    (0, 1, 2),  # Edge from 0 to 1 with weight 2
    (0, 2, 4),  # Edge from 0 to 2 with weight 4
    (1, 2, 1),  # Edge from 1 to 2 with weight 1
    (1, 3, 7),  # Edge from 1 to 3 with weight 7
    (2, 3, 3),  # Edge from 2 to 3 with weight 3
    (2, 4, 5),  # Edge from 2 to 4 with weight 5
    (3, 4, 6),  # Edge from 3 to 4 with weight 6
    (3, 5, 8),  # Edge from 3 to 5 with weight 8
    (4, 5, 9)   # Edge from 4 to 5 with weight 9
]

graph = Graph(nodes, edges)

# Run Kruskal's algorithm
mst = graph.kruskal()

# Print the Minimum Spanning Tree (MST)
print("Minimum Spanning Tree (MST):")
for u, v, w in mst:
    print(f"Edge ({u}, {v}) with weight {w}")

# Draw the original directed graph
print("Original Directed Graph:")
graph.draw_digraph()  # If you want to visualize the MST separately

# Create and visualize the solution graph (MST)
solution_graph = Graph(nodes, mst)
print("MST Visualization:")
solution_graph.draw_graph()
```
